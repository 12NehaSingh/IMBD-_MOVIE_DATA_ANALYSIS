# 17. Classify Movies based on Rating(Excellent,Good,Average,Need Improvement)

```python
def rating1(rating) :
    if rating>7.0 :
        return "Excellent"
    elif rating > 6.0 :
        return "Good"
    elif rating > 5.0:
        return "Average"
    else :
        return "Need Improvement"
data['Rating category']=data['Rating'].apply(rating1)
```
